```python
all_data['Hour']=pd.to_datetime(all_data['Order Date']).dt.hour
```

```python
keys=[]
hour=[]
for k,h_df in all_data.groupby('Hour'):
  keys.append(k)
  hour.append(len(h_df))
```

```python
all_data.groupby('Product')['Quantity Ordered'].sum().plot(kind='bar')
```

```python
all_data.groupby('Product')['Price Each'].mean()
```

```python
prod = all_data.groupby('Product')['Quantity Ordered'].sum().index
qty = all_data.groupby('Product')['Quantity Ordered'].sum()
price = all_data.groupby('Product')['Price Each'].mean()
```

```python
plt.figure(figsize=(40,24))
fig,ax1=plt.subplots()
ax2=ax1.twinx()
ax1.bar(prod,qty,color='g')
ax2.plot(prod,price,'b-')
ax1.set_xticklabels(prod,rotation='vertical',size=8)
```

```python
#Which product are most often sold together? note:- keep orders that have same order id are sold mostly together.
df=all_data[all_data['Order ID'].duplicated(keep=False)]
df.head(20)
```

```python
df['Grouped']=df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
```

```python
#The shape of a DataFrame is a tuple of array dimensions that tells the number of rows and columns of a given DataFrame.
df.shape
```

```python
#lets drop out all duplicate Order ID
df2 = df.drop_duplicates(subset=['Order ID'])
```

```python
df2['Grouped'].value_counts()[0:5].plot.pie()
```

```python
values=df2['Grouped'].value_counts()[0:5]
labels=df['Grouped'].value_counts()[0:5].index
```

```python
trace=go.Pie(labels=labels,values=values,
             hoverinfo='label+percent',textinfo='value',
             textfont=dict(size=25),
             pull=[0,0,0,0.2,0]
             )
```
